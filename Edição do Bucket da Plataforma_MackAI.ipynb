**Blob** é uma sigla para **Binary Large Object** (“objeto binário grande”).
No contexto do Google Cloud Storage, um **blob é simplesmente um arquivo armazenado dentro de um bucket**.

# Conectar ao bucket

Tenho o arquivo "key.json" na pasta raiz do seu google colab.

```python
from google.colab import files
from google.cloud import storage

client = storage.Client.from_service_account_json("key.json")
bucket = client.bucket("plataforma-mackai")
```

# Listar blobs

```python
for blob in bucket.list_blobs():
    print(blob.name)
```

# Abrir um blob

Abrir e printar

```python
blob = bucket.blob("teste_arquivo.py")
conteudo = blob.download_as_text()

print(conteudo)
```

Baixar em arquivo

```python
blob = bucket.blob("encontro/agendar.py")
blob.download_to_filename("teste_arquivo.py")
```

# Criar ou sobrescrever um blob

Direto da variável

```python
codigo = """
print("Olá do Colab!")
"""

blob = bucket.blob("colab_test.py")
blob.upload_from_string(codigo, content_type="text/x-python")
```

Do arquivo

```python
blob = bucket.blob("teste_arquivo.py")
blob.upload_from_filename("teste_arquivo.py")
```

# Apagar blob

```python
blob = bucket.blob("teste_arquivo.py")
blob.delete()
```
